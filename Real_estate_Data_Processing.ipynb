```
# Reading text file.
rdd = sc.textFile("/FileStore/tables/Practice_Data/input_real_estate.txt")
rdd.collect()
```

```
# Find headers from rdd
header_rdd = rdd.filter(lambda x: x.startswith('Property_ID') )
header_rdd.collect()

# Find data from rdd
data_rdd = rdd.filter(lambda x: not x.startswith('Property_ID'))
data_rdd.collect()
```

```
# Apply map function to find column data.
data_rdd1 = data_rdd.map(lambda x : x.split('|'))
data_rdd1.collect()
```

```
# Find header index
cols_list = header_rdd.first().split("|")
col1 = cols_list.index('Property_ID')
col2 = cols_list.index('Location')
col3 = cols_list.index('Size')
col4 = cols_list.index('Price_SQ_FT')
```

```
# funcation to calculate final price.
def calculate_final_price(Size,Price_SQ_FT):
  final_price = float(Size) * float(Price_SQ_FT)
  return str(final_price)
```

```
# Create final data headers rdd.
final_header = header_rdd.map(lambda x: x.split('|')[col1]+"|"+x.split("|")[col2]+"|Final_Price")
final_header.collect()
# Calculate final data rdd.
final_data = data_rdd1.map(lambda x:x[col1]+"|"+x[col2]+"|"+ calculate_final_price(x[col3],x[col4]))
final_data.collect()
```

```
# Merge final header and final data rdds.
final_rdd = final_header.union(final_data)
final_rdd.collect()
```
