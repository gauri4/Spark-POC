```
Read_file_df = spark.read.options(header='True',delimter=',').csv("/FileStore/tables/Practice_Data/Comma_delimeter_csv.csv")
Read_file_df.display()
```

```
Read_file_df.printSchema()
```

```
from pyspark.sql.types import IntegerType
from pyspark.sql.functions import col
Cast_df = Read_file_df.withColumn("Marks",col("Marks").cast("int"))
Cast_df.printSchema()
```

```
Transposed_df = Cast_df.groupBy('Roll_no').pivot('Subject').max("Marks")
Transposed_df.show()
```

```
Add_total_col_df = Transposed_df.withColumn("Total", Transposed_df["English"]+Transposed_df["Maths"]+Transposed_df["Physics"]+Transposed_df["Science"])
Add_total_col_df.display()
```
