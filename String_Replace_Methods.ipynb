```
df1 = spark.read.option("header", True).csv("/FileStore/tables/Practice_Data/personal_transactions.csv")
df1.display()
```

# First Method : using replace method

```
#df.na.replace("old value", "New value")
df1_replaced = df1.na.replace("Checking","Cash")
df1_replaced.display()
```

# Second method: using regexp_replace

```
from pyspark.sql.functions import regexp_replace
df2_replaced = df1.withColumn("Card_type", regexp_replace("Card_type", "Checking", "Cash"))
df2_replaced.display()
```

# Third method: using when case

```
from pyspark.sql.functions import when,col,lit
df3_replaced = df1.withColumn("Card_type", when(col("Card_type").rlike("Checking"), lit("Cash")).otherwise(col("Card_type")))
df3_replaced.display()
                              
```
